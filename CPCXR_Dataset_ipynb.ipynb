```python
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
url = 'https://raw.githubusercontent.com/ValenBte/DIALUNG/main/Images/Test/'
```

```python
df = pd.read_csv('test.csv')
#print(df)
image_number = 40
img_sel=df['filename'][image_number]
response = requests.get(url+img_sel)
#print(df)
img = Image.open(BytesIO(response.content))
diagnostico = ['Normal', 'COVID-19', 'Neumon√≠a', 'Tuberculosis']
print(diagnostico[df['label'][image_number]])

plt.xticks([])
plt.yticks([])
plt.imshow(img, cmap="gray")
```

```python
figure = plt.figure(figsize=(8, 8))
cols, rows = 3,3
total_datos = len(df.filename)
for i in range(1, cols * rows + 1):
    #sample_idx = [1,2,3,4,5,6,7,8,9]
    sample_idx = int(random.random()*total_datos)
    img_sel=df["filename"][sample_idx]#Obtener nuevamente el request de la imagen
    #print(img_sel)
    response = requests.get(url+img_sel)
    img = Image.open(BytesIO(response.content))
    #diagnostico = ['Normal', 'COVID-19', 'Neumon√≠a', 'Tuberculosis']
    label = diagnostico[df["label"][sample_idx]]

    
    figure.add_subplot(rows, cols, i)
    plt.title(label)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
plt.show()
```
