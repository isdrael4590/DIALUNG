```python
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
url = 'https://raw.githubusercontent.com/ValenBte/DIALUNG/main/Images/Test/'
```

```python
df = pd.read_csv('test.csv')
#print(df)
image_number = 40
img_sel=df['filename'][image_number]
response = requests.get(url+img_sel)
#print(df)
img = Image.open(BytesIO(response.content))
diagnostico = ['Normal', 'COVID-19', 'Pneoumonia', 'Tuberculosis']
print(diagnostico[df['label'][image_number]])

plt.xticks([])
plt.yticks([])
plt.imshow(img, cmap="gray")
```
